```python
import meep as mp
import numpy as np
import matplotlib.pylab as plt
import h5py
import math
```

## 1. Set simulation parameter

```python
# simulation resolution
resolution_um = 25
# thickness of perfect matched layer
pml_um = 1
# thickness of doe's substrate
substrate_um = 2
# operation frequency
wavelength_um = 1
frequency = 1 / wavelength_um
frequency_width = 0.2 * frequency

# simulation area
# the dimension of doe
D_um = 80
focal_length_um = 200

# Define the space (cell) for simulation¶
# for doe: wavelength and unit size is scalable
cell = mp.Vector3(D_um + 2*pml_um, D_um + 2*pml_um, focal_length_um + 2*pml_um)
```

```python
help(mp.GeometricObject)
```

# define a Gaussian beam

```python
s = 14
resolution = 80
dpml = 1

cell_size = mp.Vector3(s, s)

boundary_layers = [mp.PML(thickness=dpml)]

beam_x0 = mp.Vector3(0, 0)  # beam focus (relative to source center)
rot_angle = 0  # CCW rotation angle about z axis (0: +y axis)
beam_kdir = mp.Vector3(0, 1, 0).rotate(
    mp.Vector3(0, 0, 1), math.radians(rot_angle)
)  # beam propagation direction
beam_w0 = 2  # beam waist radius
beam_E0 = mp.Vector3(0, 0, 1)
fcen = 1
sources = [
    mp.GaussianBeamSource(
        src=mp.ContinuousSource(fcen),
        center=mp.Vector3(0, -0.5 * s + dpml + 1.0),
        size=mp.Vector3(s),
        beam_x0=beam_x0,
        beam_kdir=beam_kdir,
        beam_w0=beam_w0,
        beam_E0=beam_E0,
    )
]

sim = mp.Simulation(
    resolution=resolution,
    cell_size=cell_size,
    boundary_layers=boundary_layers,
    sources=sources,
)

sim.run(until=20)
```

```python
sim.plot2D(
    fields=mp.Ez,
    output_plane=mp.Volume(
        center=mp.Vector3(), size=mp.Vector3(s, s)
    ),
)

plt.savefig(f"Ez_angle{rot_angle}.png", bbox_inches="tight", pad_inches=0)
```
